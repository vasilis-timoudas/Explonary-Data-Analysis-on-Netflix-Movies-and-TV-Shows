###  4.2 Drop duplicate values 
<font color='black'>The second thing that I will do is to drop duplicate values. We see that there's 0 duplicated values.</font> <a class="anchor" id="chapter4_2">

```python
# Count the duplicate values
duplicate_values_count = netflix_data.duplicated().sum()

# Print the count of the duplicate values
print("Count of duplicate values before dropping: ", duplicate_values_count)

# Drop duplicate values 
netflix_data = netflix_data.drop_duplicates()

# Count the duplicate values
duplicate_values_count = netflix_data.duplicated().sum()

# Print the count of the duplicate values
print("Count of duplicate values after dropping : ", duplicate_values_count)
```

###  4.3 Convert date to datetime
<font color='black'>The third thing I will do is to convert the date column <b>date_added</b> to datetime.</font>
<a class="anchor" id="chapter4_3">

```python
# Print the head of the dataframe in the column date_added
print("Before parsing dates:\n", netflix_data['date_added'].head())

# Parsing dates
netflix_data['date_added'] = pd.to_datetime(netflix_data['date_added'])

# Print the head (first 5 rows) of the dataframe in the column date_added 
print("\nAfter parsing dates:\n", netflix_data['date_added'].head() )
```

### 4.4 Add three new columns for year, month and day 
<font color='black'>Create three new columns for the year, month and day added.</font> <a class="anchor" id="chapter4_4">

```python
# Add 3 new columns for the year, month and day seperetaly
netflix_data['year_added'] = netflix_data['date_added'].dt.year
netflix_data['month_added'] = netflix_data['date_added'].dt.month
netflix_data['day_added'] = netflix_data['date_added'].dt.day
```

### 4.5 Add a new column for the season 
<font color='black'>From the column 'month_added' create a new column for the season.</font> <a class="anchor" id="chapter4_5">

```python
# Map for seasons
season_map = {12: 1,
              1:  1,
              2:  1,
              3:  2,
              4:  2,
              5:  2,
              6:  3,
              7:  3,
              8:  3,
              9:  4,
              10: 4,
              11: 4}

# Create a new column for the season
netflix_data['season_added'] = netflix_data['month_added'].map(season_map)
```

### 4.6 Add a new column that will contain the principal country
<font color='black'>We notice that the column <b>country</b> contains multiple countries. So I will create a new column <b>principal_country</b> that will have the principal (first) country of the column <b>country</b>.</font> <a class="anchor" id="chapter4_6">

```python
# Print the rows 7-13 of the column country
print(netflix_data['country'][7:13])

# Create a new column and put the store the first country of the column country
netflix_data['principal_country'] = netflix_data['country'].str.split(',').str[0].str.strip()

# Print the rows 7-13 of the column country
print("\n",netflix_data['principal_country'][7:13])
```

### 4.7 Add a new column that will contain the age groups
<font color='black'>We noticed that there's a column called <b>rating</b> that contains maturity ratings. I will create a new column called <b>maturity</b> with the values 'Kids', 'Teens', or 'Adults' so for every Movie or TV Show in the dataframe, we will know the appropriate age group. To do that I used the fuction <b>maturity_ratings</b> which I made from the website https://help.netflix.com/en/node/2064/us and will convert manurity ratings into age groups.</font> <a class="anchor" id="chapter4_7">

```python
# Fuction that converts maturity ratings into age groups
def maturity_ratings(rating):
    if rating in ['TV-Y', 'TV-Y7', 'G', 'TV-G', 'PG', 'TV-PG']:
        return 'Kids'
    elif rating in ['PG-13', 'TV-14']:
        return 'Teens'
    elif rating in ['R', 'TV-MA', 'NC-17']:
        return 'Adults'

# Print the head of the dataframe in the column rating 
print(netflix_data['rating'].head(), "\n")

# Create the new column maturity from the function maturity_ratings
netflix_data['maturity'] = netflix_data['rating'].apply(maturity_ratings)

# Print the head of the dataframe in the column maturity 
print(netflix_data['maturity'].head())
```

### 4.8 Strip and replace the white spaces
<font color='black'>Afterwards I will strp the white spaces in the column</font> <a class="anchor" id="chapter4_8">

```python
# Strip white spaces
netflix_data['type'] = netflix_data['type'].str.strip()
netflix_data['rating'] = netflix_data['rating'].str.strip()
```

<font color='black'>I will replace the white spaces between commas in the columns</font>

```python
# Replace the white spaces between commas
netflix_data['director'] = netflix_data['director'].str.replace(r'\s*,\s*', ',', regex=True)
netflix_data['cast'] = netflix_data['cast'].str.replace(r'\s*,\s*', ',', regex=True)
netflix_data['country'] = netflix_data['country'].str.replace(r'\s*,\s*', ',', regex=True)
netflix_data['listed_in'] = netflix_data['listed_in'].str.replace(r'\s*,\s*', ',', regex=True)
```

### 4.9 Handle the missing values
<font color='black'>We notice that the columns <b>director</b>, <b>cast</b>, <b>country</b>, <b>date_added</b>, <b>rating</b> and <b>duration</b> have missing values. The columns <b>director</b>, <b>cast</b> and <b>country</b> have the most missing values. <br>I have two choices, to drop the missing values or to keep and fill them. I decided that the best choice is to fill them because maybe they can be critical for the analysis.</font> <a class="anchor" id="chapter4_9">

```python
# Count the missing values
missing_values_count = netflix_data.isnull().sum()

# Print the count of the missing values
print("Missing values before filling count:")
print(missing_values_count) 

# Fill missing values with 'Unknown'
netflix_data['director'].fillna('Unknown', inplace=True)
netflix_data['cast'].fillna('Unknown', inplace=True)
netflix_data['country'].fillna('Unknown', inplace=True)
netflix_data['date_added'].fillna('Unknown', inplace=True)
netflix_data['rating'].fillna('Unknown', inplace=True)
netflix_data['duration'].fillna('Unknown', inplace=True)
netflix_data['principal_country'].fillna('Unknown', inplace=True)
netflix_data['maturity'].fillna('Unknown', inplace=True)
netflix_data['year_added'].fillna('Unknown', inplace=True)
netflix_data['month_added'].fillna('Unknown', inplace=True)
netflix_data['day_added'].fillna('Unknown', inplace=True)
netflix_data['season_added'].fillna('Unknown', inplace=True)
```

```python
missing_values_count = netflix_data.isnull().sum()
print("Missing values after filling count:")
print(missing_values_count) 
```

### 5.3 Plot the top 10 countries of Movies & TV Shows split 
<font color='black'>I will plot the top 10 countries Movies & TV Shows split. As we saw earlier the countries are the same as before but now I will plot also the split between Movies and TV Shows. We see that most countries they prefer Movies instead of TV Shows.</font> <a class="anchor" id="chapter5_3">

```python
# Create a cross-tabulation of the columns country and type 
country_type_counts = pd.crosstab(netflix_data['principal_country'], netflix_data['type'])

# Create a new column 'total' that adds each row on axis 1 
country_type_counts['total'] = country_type_counts.sum(axis=1)

# Sort and then reverse and select the top 10 countries (ignore 'Unknown') with the highest value in column 'total'
top10_country_type_counts = country_type_counts[country_type_counts.index != 'Unknown'].sort_values(by='total').iloc[::-1][:10]

print(top10_country_type_counts)
```

```python
# Reverse the dataframe 
top10_country_type_counts = top10_country_type_counts.iloc[::-1] # Sometimes this line needs so you can show the bar plot reversed

# Create a figure and a set of subplots
fig, ax = plt.subplots(1, 1, figsize=(18, 9))

# Create a horizontal bar plot for the 'Movie'
ax.barh(top10_country_type_counts.index, 
        top10_country_type_counts['Movie'], 
        color=netflix_palette[1], 
        alpha=0.8, 
        label='Movie')

# Create a horizontal bar plot for the 'TV Show'
ax.barh(top10_country_type_counts.index, 
        top10_country_type_counts['TV Show'], 
        color=netflix_palette[0], 
        alpha=0.8, 
        label='TV Show',
        left=top10_country_type_counts['Movie'])

# Set ticks on y-axis 
ax.set_yticks(range(len(top10_country_type_counts.index)));

# Set tick labels on y-axis 
ax.set_yticklabels(top10_country_type_counts.index, fontfamily='serif', fontsize=11);

# Initialize fontsize and fontweights
fontsize = 7
fontweight_movie = 'light'
fontweight_tvshow = 'light'

# Set annotations
for i in top10_country_type_counts.index:
    # Bold if the value of Movie is higher than the value of TV Show
    if top10_country_type_counts['Movie'][i] > top10_country_type_counts['TV Show'][i]:
        fontweight_movie = 'bold'
        fontweight_tvshow = 'light'
    # Bold if the value of TV Show is higher or equal to the value of Movie
    else:
        fontweight_movie = 'light'
        fontweight_tvshow = 'bold'
    
    # Annotate Movie
    ax.annotate(f"{top10_country_type_counts['Movie'][i]/top10_country_type_counts['total'][i]*100:.1f}%", 
                   xy=(top10_country_type_counts['Movie'][i]/2, i),
                   va='center', 
                   ha='center',
                   fontsize=fontsize, 
                   fontweight=fontweight_movie, 
                   fontfamily='serif', 
                   color='white')
    
    # Annotate TV Show
    ax.annotate(f"{top10_country_type_counts['TV Show'][i]/top10_country_type_counts['total'][i]*100:.1f}%", 
                   xy=(top10_country_type_counts['Movie'][i]+top10_country_type_counts['TV Show'][i]/2, i),
                   va='center',
                   ha='center', 
                   fontsize=fontsize, 
                   fontweight=fontweight_tvshow, 
                   fontfamily='serif', 
                   color='white')
    
    # Increase fontsize by 0.25
    fontsize = fontsize + 0.25 

# Remove the border from the plot
for s in ['top', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Set grid on x-axis
ax.grid(axis='x', linestyle='-', alpha=0.4) 
    
# Set ticks on x-axis 
ax.set_xticks(np.arange(0, 4000, 500));

# Set texts
fig.text(0.75, 0.88, "Movie", fontweight="bold", fontfamily='serif', fontsize=16, color=netflix_palette[1]);
fig.text(0.80, 0.88, "|", fontweight="bold", fontfamily='serif', fontsize=16, color='black');
fig.text(0.82, 0.88, "TV Show", fontweight="bold", fontfamily='serif', fontsize=16, color=netflix_palette[0]);

# Set the title of the horizontal bar
plt.title('Top 10 countries Movie & TV Show split', fontsize=20, fontweight='bold', fontfamily='serif')

# Show the horizontal bar
plt.show()
```

### 5.8 Plot the rating rating distribution by Movies and TV Shows 
<font color='black'>Now I will plot the rating distribution by Move & TV Show. But first of all i've noticed a problem. Three rows don't have their values correctly. We see that the column <b>rating</b> have the values of the column <b>duration</b>. To solve this problem I will use the function <b>swap_columns_values</b>. This function will swap the values of 2 columns. The TV-MA maturity rating tends to be the preffered choicea among most people.</font> <a class="anchor" id="chapter5_8">

```python
# USA maturity ratings on Netflix
ratings = ['TV-Y', 'TV-Y7', 'G', 'TV-G', 'PG', 'TV-PG', 'PG-13', 'TV-14', 'R', 'TV-MA', 'NC-17', 'NR', 'UR', 'TV-Y7-FV']

# In column rating find the values that are not in list ratings and ignore 'Unknown'
rows_not_in_rating = netflix_data['rating'][~netflix_data['rating'].isin(ratings) & (netflix_data['rating'] != 'Unknown')]
print(rows_not_in_rating)

# Print the 3 rows in the 3 columns 
selected_rows = netflix_data.loc[[5541, 5794, 5813], ['type', 'rating', 'duration']]
selected_rows
```

```python
# Function to swap the values of 2 columns
def swap_columns_values(netflix_data, column_rating='rating', column_duration='duration', ratings=ratings):
    for i in range(len(netflix_data)):
        # Get the rating value of the row in index i
        row_rating = netflix_data.loc[i, column_rating]
        
        # Check if the duration value of row in index i is empty
        row_duration_is_empty = pd.isna(netflix_data.loc[i, column_duration])
        
        # If row_rating value is not in list rating and the row_duration_is_empty value is not empty 
        if row_rating not in ratings and not row_duration_is_empty:
            # Swap the values of the columns duration and rating
            netflix_data.loc[i, column_duration] = netflix_data.loc[i, column_rating]
            netflix_data.loc[i, column_rating] = 'Unknown'
    
    # Return dataframe
    return netflix_data
    
# Renew the netflix dataframe
netflix_data = swap_columns_values(netflix_data)
```

```python
# Group the columns 'rating' and 'type' (ingore when rating is 'Unknown'), count the values of column 'type' and unstack the dataframe
type_ratings_counts = netflix_data[netflix_data['rating'] != 'Unknown'].groupby('rating')['type'].value_counts().unstack(fill_value=0)

# Create new column 'Total' that has the sum of the axis=1 
type_ratings_counts['Total'] = type_ratings_counts.sum(axis=1)

# Sort by the column 'Total'
type_ratings_counts = type_ratings_counts.sort_values(by='Total')[::-1]

print(type_ratings_counts)

# In column 'TV Show' convert the values to negatives
type_ratings_counts['TV Show'] = -type_ratings_counts['TV Show']
```

```python
# Create a figure and a set of subplots
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# Create two bar plots
ax.bar(type_ratings_counts.index, type_ratings_counts['Movie'], width=0.5, color=netflix_palette[1], alpha=0.8, label='Movie')
ax.bar(type_ratings_counts.index, type_ratings_counts['TV Show'], width=0.5, color=netflix_palette[0], alpha=0.8, label='TV Show')

# Draw a line at point y=0
ax.axhline(y=0, color=netflix_palette[0], linestyle='--')

# Set annotation for Movie
for i in type_ratings_counts['Movie'].index:
    # Annotate Movie
    ax.annotate(f"{type_ratings_counts['Movie'][i]}", 
                   xy=(i, type_ratings_counts['Movie'][i] + 100),
                   va = 'center', 
                   ha='center', 
                   fontweight='light', 
                   fontfamily='serif',
                   color='#4a4a4a')   

# Set annotation for TV Show
for i in type_ratings_counts['TV Show'].index:
    # Annotate TV Show
    ax.annotate(f"{-type_ratings_counts['TV Show'][i]}", 
                   xy=(i, type_ratings_counts['TV Show'][i] - 100),
                   va = 'center', 
                   ha='center',
                   fontweight='light', 
                   fontfamily='serif',
                   color='#4a4a4a')

# Remove the border from the plot
for s in ['top', 'right']:
    ax.spines[s].set_visible(False)

# Set texts
fig.text(0.755,0.924,"Movie", fontweight="bold", fontfamily='serif', fontsize=15, color=netflix_palette[1])
fig.text(0.815,0.924,"|", fontweight="bold", fontfamily='serif', fontsize=15, color='black')
fig.text(0.825,0.924,"TV Show", fontweight="bold", fontfamily='serif', fontsize=15, color=netflix_palette[0])

# Set title for the two bar plots
plt.title('Rating distribution by Movie & TV Show', fontsize=20, fontweight='bold', fontfamily='serif', y=1.1)

# Show the two bar plots
plt.show()
```

<font color='black'> Function to calculate the top10 countries difference between added and released year.</font>

```python
# Calculate difference between added and released year for the top 10 countries on Netflix
def countries_years_diff(_type):
    # Select rows from netflix_data based on specific conditions and keep only the following columns
    type_countries_years = netflix_data.loc[(netflix_data['type'] == _type) & 
                                              (netflix_data['year_added'] != 'Unknown') & 
                                              (netflix_data['principal_country'] != 'Unknown'), 
                                            ['principal_country', 'release_year', 'year_added']].copy()
    
    # Group the above selected rows and round the values of the columns release_year and year_added (keep only the last 2 digits)
    countries_years = type_countries_years.groupby('principal_country').agg(
                                            {'release_year': lambda x: round(x.mean(), 2), 
                                             'year_added': lambda x: round(x.mean(), 2)})
    
    # Filtering the above dataframe to include only the top 10 countries
    top10_countries_years_diff = countries_years.loc[countries_years.index.isin(top10_countries.index)].copy()

    # Calculate the difference
    top10_countries_years_diff['years_difference'] = top10_countries_years_diff['year_added'] - top10_countries_years_diff['release_year']

    # Round the difference and keep only the last digit
    top10_countries_years_diff['years_difference'] = top10_countries_years_diff['years_difference'].round(1)
    
    # 
    top10_countries_years_diff = top10_countries_years_diff.sort_values(by='years_difference')[::-1]
    
    # Return the 
    return top10_countries_years_diff
```

<font color='black'> Function to plot the difference between added and released year.</font>

```python
# Plot difference between added and release year
def plot_years_distance(type_years_diff):
    
    # Range from 1 until the length of type_years_diff plus 1
    my_range = range(1, len(type_years_diff.index)+1)

    # Create a figure and a set of subplots
    fig, ax = plt.subplots(1, 1, figsize=(7, 5))

    # Create a hline
    ax.hlines(y=my_range, 
               xmin=type_years_diff['release_year'], 
               xmax=type_years_diff['year_added'], 
               color='grey', 
               alpha=0.4)

    # Create a scatter for the release year
    sc1 = ax.scatter(type_years_diff['release_year'], 
                my_range, 
                color=netflix_palette[0],
                s=100, 
                alpha=0.9, 
                label='Average release date')

    # Create a scatter for the year added
    sc2 = ax.scatter(type_years_diff['year_added'],
                my_range, 
                color=netflix_palette[1],
                s=100, 
                alpha=0.9 , 
                label='Average added date')

    # Remove the border from the plot
    for s in ['top', 'left']:
        ax.spines[s].set_visible(False)

    # Modifications applied to both y and x axis
    ax.tick_params(axis=u'both', which=u'both',length=0)

    # Move tick marks to the right side of the plot
    ax.yaxis.tick_right()

    # Set ticks on y-axis 
    plt.yticks(my_range, type_years_diff.index, fontname = "serif",fontsize=11)

    # Set tick labels on the x-axis
    ax.set_xticks(np.arange(2011, 2021, 1))

    # Set texts
    fig.text(0.19, 0.175,"Released", fontweight="bold", fontfamily='serif', fontsize=11, color=netflix_palette[0])
    fig.text(0.76, 0.175,"Added", fontweight="bold", fontfamily='serif', fontsize=11, color=netflix_palette[1])

    # Set text on hline and between the 2 scatters
    for i, (x1, x2) in enumerate(zip(sc1.get_offsets(), sc2.get_offsets())):
        # Years text
        text = str(type_years_diff['years_difference'][i]) + ' years'

        # if the years are less or equal to 1 then the text will contain the word 'year'
        if type_years_diff['years_difference'][i] <= 1:
            text = str(type_years_diff['years_difference'][i]) + ' year'

        # Mid point between the 2 scatters
        mid_point = np.mean([x1, x2], axis=0)

        # Set text
        ax.text(mid_point[0], mid_point[1]+0.25, text, ha='center', va='center', color=netflix_palette[0], fontfamily='serif', fontsize=11)

    # Set the title
    plt.title('Distance between movies release and added date on Netflix', fontsize=12, fontweight='bold', fontfamily='serif', y=1.02)

    # Show the hline and scatters
    plt.show()
```

<font color='black'> From now on I decided that is good idea to not compare only countries but to compare also regions, specific parts of a region and languages. I used the function below to make a dataframe that will contain the content that added each year.</font>

```python
# Added content each year
def region_content_added(region_list, region_name, countries_content_added=countries_content_added):
    
    # Filter the countries_content_added based on the region_list
    region_df = countries_content_added[countries_content_added['principal_country'].isin(region_list)].copy()
    
    # Ignore column 'Uknown'
    region_df = region_df[region_df['year_added'] != 'Unknown']
    
    # Group the column 'year_added' and sum the numeric values in each group
    region_df = region_df.groupby('year_added').sum(numeric_only=True).reset_index()[['year_added', 'count']]
    
    # Add a new column 'region' with the specified region_name
    region_df['region'] = region_name
    
    # Reorder the columns of the DataFrame
    region_df = region_df.reindex(columns=['region', 'year_added', 'count'])
    
    # Return region's dataframe
    return region_df
```

<font color='black'> I used the function below to plot two contents that were added to Netflix so that I can compare them.</font>

```python
# Plot two contents
def plot_content_comparison(contentA, contentB, x, y=[0.2, 0.2, 0.2], fontsize=20):
    
    # Sum the column 'count' of the 2 contents
    sum_contentA = contentA['count'].sum()
    sum_contentB = contentB['count'].sum()
    
    # Calculate and find the minimum year in the column 'year_added' of the 2 contents
    min_year_contentA = contentA['year_added'].min()
    min_year_contentB = contentB['year_added'].min()
    min_year_content = min(min_year_contentA, min_year_contentB)
    
    # Calculate and find the maximum year in the column 'year_added' of the 2 contents
    max_year_contentA = contentA['year_added'].max()
    max_year_contentB = contentB['year_added'].max()
    max_year_content = max(max_year_contentA, max_year_contentB)
    
    
    # Calculate and find the maximum year in the column 'count' of the 2 contents
    max_count_contentA = contentA['count'].max()
    max_count_contentB = contentB['count'].max()
    max_count_content = max(max_count_contentA, max_count_contentB)
    
    # Set the content colors
    color_contentA = netflix_palette[0] # Black color
    color_contentB = netflix_palette[1] # Red color
    
    # Create a figure and a set of subplots
    fig, ax = plt.subplots(1, 1, figsize=(12, 6))
    
    # I want the content with the highest sum to have the black color
    # If the sum of the contentA is bigger than the sum of the contentB, 
    # then the contentA will have the black color and the contentB will have the red color
    if(sum_contentA > sum_contentB):
        # Plot the contentA
        ax.plot(contentA['year_added'], 
                contentA['count'],
                color=color_contentA,
                label=contentA['region'][0])
        
        # Fill the plot of the contentA
        ax.fill_between(contentA['year_added'],
                        0,
                        contentA['count'],
                        color=color_contentA,
                        alpha=0.9)
        
        # Plot the contentB
        ax.plot(contentB['year_added'], 
                contentB['count'],
                color=color_contentB,
                label=contentB['region'][0])

        # Fill the plot of the contentB
        ax.fill_between(contentB['year_added'],
                        0,
                        contentB['count'],
                        color=color_contentB,
                        alpha=0.9)
        
    # # Else if the sum of the contentB is bigger than the sum of the contentA, 
    # then the contentB will have the black color and the contentA will have the red color
    else:
        # Set the content colors 
        color_contentB = netflix_palette[0] # Black color
        color_contentA = netflix_palette[1] # Red color
        
        # Plot the contentB
        ax.plot(contentB['year_added'], 
                contentB['count'],
                color=color_contentB,
                label=contentB['region'][0])

        # Fill the plot of the contentB
        ax.fill_between(contentB['year_added'],
                        0,
                        contentB['count'],
                        color=color_contentB,
                        alpha=0.9)
        
        # Plot the contentA
        ax.plot(contentA['year_added'], 
                contentA['count'],
                color=color_contentA,
                label=contentA['region'][0])

        # Fill the plot of the contentA
        ax.fill_between(contentA['year_added'],
                        0,
                        contentA['count'],
                        color=color_contentA,
                        alpha=0.9)
        
    # Remove the border from the plot
    for s in ['top', 'left']:
        ax.spines[s].set_visible(False)
        
    # Move tick marks to the right side of the plot 
    ax.yaxis.tick_right()

    # Set x and y margins to 0
    ax.margins(x=0)
    ax.margins(y=0)
    
    # Modifications applied to both y and x axis
    ax.tick_params(axis=u'both', which=u'both',length=0)
    
    # Set tick labels on the x-axis and y-axis
    plt.xticks(np.arange(min_year_content, max_year_content+1, 1), fontfamily='Serif')
    plt.yticks(np.arange(0, max_count_content+200, 200), fontfamily='Serif')
        
    # Set texts
    fig.text(x[0], y[0], contentA['region'][0], fontweight="bold", fontfamily='serif', fontsize=15, color=color_contentA)
    fig.text(x[1], y[1], "|", fontweight="bold", fontfamily='serif', fontsize=15, color='black')
    fig.text(x[2], y[2],contentB['region'][0], fontweight="bold", fontfamily='serif', fontsize=15, color=color_contentB)

    # Set title
    title = 'Content Added on Netflix: ' + str(contentA['region'][0]) + ' vs ' + str(contentB['region'][0])
    plt.title(title, fontsize=fontsize, fontweight='bold', fontfamily='serif')    
        
   # plt.legend()
    
    # Plot the 2 contents
    plt.show()
```

```python
# Create a heatmap for maturity ratings
def maturity_ratings_heatmap(maturity_rating, age_order = ['Kids', 'Teens','Adults']):
    # Create a cross-tabulation of the columns 'principal_country' and 'maturity' and transpose the matrix
    rating_heatmap = pd.crosstab(maturity_rating['principal_country'],
                             maturity_rating['maturity'],
                             normalize = "index").T
    
    # Drop the row 'Unknown'
    rating_heatmap = rating_heatmap.drop('Unknown')
    
    # Reindex the dataframe to 'age_order'
    rating_heatmap = rating_heatmap.reindex(age_order)
    
    # Return rating heatmap dataframe
    return rating_heatmap
```

```python
# Plot maturity ratings heatmap
def plot_maturity_ratings_heatmap(rating_heatmap, cmap=cmap_netflix, top10_countries=top10_countries, title='Maturity ratings for Movies on Netflix'):
    # Create a figure
    fig, ax = plt.subplots(1, 1, figsize=(12, 12))
    
    # Plot heatmap
    sns.heatmap(rating_heatmap.loc[:, top10_countries.index],
            cmap=cmap_netflix,
            square=True, 
            linewidth=2.5,
            cbar=False,
            annot=True,
            fmt='1.0%',
            vmax=.6,
            vmin=0.05,
            ax=ax,
            annot_kws={"fontsize":12})
    
    # Set title
    plt.title(title, fontsize=20, fontweight='bold', fontfamily='serif', y=1.02)

    # Show the heatmap
    plt.show()
```
